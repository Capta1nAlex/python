```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data = np.random.randint(-10, 11, size=(5, 5))
sns.heatmap(data, annot=True, cmap='coolwarm', vmin=-10, vmax=10)
plt.title('Heatmap of 5x5 Array')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
