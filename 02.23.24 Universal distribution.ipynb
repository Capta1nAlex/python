```python
import numpy as np
import matplotlib.pyplot as plt
dice = np.random.randint(1, 7, 300)
hist = np.histogram(dice, bins=[1,2,3,4,5,6,7,])
labels = ['1', '2', '3', '4', '5', '6']
plt.bar(labels, hist[0])
plt.xlabel('Number rolled')
plt.ylabel('Roll count')
plt.title('Bar chart of dice rolls')




```
