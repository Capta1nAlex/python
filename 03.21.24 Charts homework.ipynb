```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,300,10).astype(np.int16)
y = 2 * x
z = 3 * x
print(x)
print("\n")
[x,y,z]

fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})
ax1.plot(x, y)
ax2.plot(y, z)
ax1.set_title('Top Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax2.set_title('Bottom Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')


plt.subplots_adjust(hspace=0.5) # Adjusting the space, so the labels wouldn't intersect each other
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating sinus and cosinus functions
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a plot
plt.plot(x, y1, label='Sin(x)')
plt.plot(x, y2, label='Cos(x)')
plt.title('Sin(x) and Cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # Showing the legend with labels for each function

plt.show()
```
