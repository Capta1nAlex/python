```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Set the initial parametres
median_salary = 1500
lower_bound = 750
upper_bound = 2250
mu = np.log(median_salary)
sigma = (np.log(upper_bound) - np.log(lower_bound)) / (2 * 0.68)

# Generating random salaries array from the log-normal distribution
random_salaries = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

# Creating a histogram
plt.hist(random_salaries, bins=45, color='blue', edgecolor='black')

# Adding labels
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Amount of People')
plt.show()
```
